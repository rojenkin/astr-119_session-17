## Perform a simple Monte Carlo integration to compute Pi

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### Just for fun; color

```python
#use colors.subclass(or command; e.g bold).colorname to print
#examples: print(colors.bold, colors.fg.blue, "this will be bold and blue")
#everything after this will have that format until the following command
#is given: print(colors.reset, "now, this text will be normal")

class colors:
    reset='\033[0m' #reset all colors with colors.reset
    bold='\033[01m'
    underline='\033[04m'
    strikethrough='\033[09m'
    reverse='\033[07m'
    class fg:  #foreground subclass
        black='\033[30m'
        red='\033[31m'
        green='\033[32m'
        orange='\033[33m'
        blue='\033[34m'
        purple='\033[35m'
        cyan='\033[36m'
        lightgrey='\033[37m'
        darkgrey='\033[90m'
        lightred='\033[91m'
        lightgreen='\033[92m'
        yellow='\033[93m'
        lightblue='\033[94m'
        pink='\033[95m'
        lightcyan='\033[96m'
    class bg: #background subclass
        black='\033[40m'
        red='\033[41m'
        green='\033[42m'
        orange='\033[43m'
        blue='\033[44m'
        purple='\033[45m'
        cyan='\033[46m'
        lightgrey='\033[47m'
```

#### Set some parameters for the integration

```python
n = 10000    #number of samples for the integration
```

### Make some uniformly sampled variables [-1,1]

```python
x = np.random.uniform(-1,1,n)
y = np.random.uniform(-1,1,n)
```

### Find the number of samples within the unit circle

```python
ir = np.where((x**2 + y**2)<1.0)[0]
ur = np.where((x**2 + y**2)>=1.0)[0]
```

### Plot the samples and the cirlce

```python
fig = plt.figure(figsize=(7,7))
plt.xlim([-1.1,1.1])
plt.ylim([-1.1,1.1])
plt.plot(x[ir],y[ir],'.',color="magenta")
plt.plot(x[ur],y[ur],'.',color="cyan")
theta = np.linspace(0,2*np.pi,1000)
xc = np.cos(theta)
yc = np.sin(theta)
plt.plot(xc,yc,color="k")

plt.xlabel('x',color="chocolate")
plt.ylabel('y',color="chocolate")
plt.show()
```

### Report the result for Pi

```python
pi_approx = 4.0*len(ir)/float(n)
error_pi = (pi_approx-np.pi)/np.pi

print(colors.underline,colors.fg.pink,"Number of examples =",colors.reset,colors.bold,colors.fg.cyan,n,colors.reset)
print(colors.underline,colors.fg.pink,"Approximate pi     =",colors.reset,colors.bold,colors.fg.cyan,pi_approx,colors.reset)
print(colors.underline,colors.fg.pink,"Error in approx    =",colors.reset,colors.bold,colors.fg.cyan,error_pi,colors.reset)
```
